```python
# Tópico 3 da AE1
import torch

# Derivadas parciais de funções escalares com até três variáveis, explorando polinômios e trigonométricas.
import sympy as sym # Essa biblioteca permite lidar com variaveis e trabalhar com expressões geométricas sem arredonda-las.

x, y, z = sym.symbols('x y z')

# Utilizando polinômios
f = 3*x**2 + 5*y - 6*z**3
df_dx = sym.diff(f, x)
df_dy = sym.diff(f, y)
df_dz = sym.diff(f, z)

# Utilizando funções trigonométricas:
g= sym.cos(3*x + 2*y -5*z)
dg_dx = sym.diff(g, x)
dg_dy = sym.diff(g, y)
dg_dz = sym.diff(g, z)

print("Derivadas parciais do polinômio:")
print("∂f/∂x =", df_dx)
print("∂f/∂y =", df_dy)
print("∂f/∂z =", df_dz)

print("\nDerivadas parciais da trigonométrica:")
print("∂g/∂x =", dg_dx)
print("∂g/∂y =", dg_dy)
print("∂g/∂z =", dg_dz)

### Operações de cálculo vetorial e matricial:

# Produto escalar:

a = sym.Matrix([2, 3, 5])
b = sym.Matrix([4, 1, 7])

produto_escalar = a.dot(b)

print("Produto escalar =", produto_escalar)

# Produto vetorial:

produto_vetorial = a.cross(b)
print("Produto vetorial =", produto_vetorial)

# Produto matriz-vetor:

vetor = sym.Matrix([[1, 2, 3]])
print(vetor)

produto_matriz_vetor = a*vetor
print("Produto matriz-vetor =", produto_matriz_vetor)

# Produto matriz-matriz:

C = sym.Matrix([[1, -2], [4, 5]])
D = sym.Matrix([[2, 7], [1, 8]])

E= C*D
print("Produto matriz-matriz =", E)

# Cálculo de determinante:

determinante = E.det()
print("Determinante =", determinante)

# Cálculo de inversa:

inversa = E.inv()
print("Inversa =", inversa)

# Transposições:

F= E.T
print("Transposta =", F)

# Cálculo de autovalores e autovetores:

autovalores, autovetores = E.eigenvals(), E.eigenvects()
print("Autovalores =", autovalores)

print("Autovetores =", autovetores)

### Para campos escalares e vetoriais:

# gradientes:
g , h, i = sym.symbols('g h i')
k = g**2 + h**2 + i**2
Gradiente = [sym.diff(k, var) for var in (g, h, i)]

print("Gradiente =", Gradiente)

# divergentes:

div_K = sym.diff(k, g) + sym.diff(k, h) + sym.diff(k, i)
print("Divergente:", div_K)

# hessianas:

hessiana = sym.hessian(k, (g, h, i))
print("Hessiana =", hessiana)

# jacobianos:
j, k = sym.symbols('j k')
F= sym.Matrix([j**2 * k, k**2 * j])

J = F.jacobian([j, k])
print("Jacobiana simbólica:", J)

# rotacionais.
vi = sym.symbols('vi')       # estou utilizando para simbolizar o vetor i (Como eu já escrevi no código, não quis executar um erro)
F_rot = [j*k, vi*k, vi*j]

rot = [
    sym.diff(F_rot[2], j) - sym.diff(F_rot[1], k),  # Componente i (Simbolizada por vi)
    sym.diff(F_rot[0], k) - sym.diff(F_rot[2], vi),  # Componente j
    sym.diff(F_rot[1], vi) - sym.diff(F_rot[0], j)   ]# Componente k

print("Rotacional (curl) =", rot)
# Foi escolhido um campo com rotacional nulo de propósito, como voltou um vetor nulo indica que ele é irrotacional (Como o esperado)
```

```python
### Use biblioteca gráficas (matplotlib, etc) para plotar campos escalares e vetoriais: (1d, 2d, e 3d).
import numpy as np
import matplotlib.pyplot as plt

# 1D:
x = np.linspace(0, 5, 100)
f = np.sin(x)


print("1D:")
plt.figure(figsize=(8, 4))
plt.plot(x, f, label=r'$f(x) = \sin(x)$')
plt.title("Campo Escalar 1D")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()

# 2D:

y = np.linspace(0, 5, 100)
X, Y = np.meshgrid(x, y)

# Campo:

Z = np.sin(X) * np.cos(Y)

print("2D:")
plt.figure(figsize=(8, 6))
contour = plt.contourf(X, Y, Z, cmap='viridis')
plt.colorbar(contour, label='f(x, y) = sin(x) * cos(y)')
plt.title('Campo Escalar 2D')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
# 3D:

# Campo escalar 3D ( Superficie)

W = np.sin(np.sqrt(X**2 + Y**2))

# Criação de uma figura 3D
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plotar superficie:
surf = ax.plot_surface(X, Y, Z, cmap='viridis')
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5, label='f(x, y) = sin(sqrt(x^2 + y^2))')
ax.set_title('Campo Escalar 3D - Superfície')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
print("3D:")
plt.show()
```

```python
### Buscar em livros de Mecânica dos Sólidos, Mecânica do Contínuo e/ou Teoria da Elasticidade exemplos considerando: campos de deslocamento (1d, 2d e 3d):
import sympy as sym

### derivar tensor de deformação

# Definição de variáveis espaciais

x, y, z = sym.symbols('x y z')

# Definição de campos de deslocamento como funções de x, y, z
u = sym.Function('u')(x, y, z)
v = sym.Function('v')(x, y, z)
w = sym.Function('w')(x, y, z)

# Criação do vetor deslocamento
displacement = sym.Matrix([u, v, w])

# Calcular tensor de deformação infinitesimal
epsilon = sym.MutableDenseMatrix(3, 3, lambda i, j: (displacement[i].diff([x,y,z][j]) + displacement[j].diff([x,y,z][i]))/2)

print("Tensor de deformação:")
sym.pprint(epsilon)

### tensor de material:

# Índices
i, j, k, l = sym.symbols('i j k l', integer=True)
delta = sym.KroneckerDelta

# Parâmetros de Lamé
lmbda, mu = sym.symbols('lambda mu')

# Tensor deformação ε_kl
epsilon = sym.MatrixSymbol('epsilon', 3, 3)

# Tensor tensão σ_ij = λ δ_ij ε_kk + 2μ ε_ij
sigma = lmbda * delta(i, j) * sym.Trace(epsilon) + 2 * mu * epsilon[i,j]

# Exibir expressão simbólica para σ_ij
print("Lei constitutiva (tensor de material isotrópico):")
print(sigma)

### tensor de tensões:

# Definição dos símbolos para as componentes do tensor de tensões
sigma_xx, sigma_yy, sigma_zz = sym.symbols('sigma_xx sigma_yy sigma_zz')
sigma_xy, sigma_yx = sym.symbols('sigma_xy sigma_yx')
sigma_xz, sigma_zx = sym.symbols('sigma_xz sigma_zx')
sigma_yz, sigma_zy = sym.symbols('sigma_yz sigma_zy')

# Matriz do tensor de tensões
sigma = sym.Matrix([
    [sigma_xx, sigma_xy, sigma_xz],
    [sigma_yx, sigma_yy, sigma_yz],
    [sigma_zx, sigma_zy, sigma_zz]
])

# Se quiser forçar a simetria (tensor simétrico)
sigma_sym = sym.Matrix([
    [sigma_xx, sigma_xy, sigma_xz],
    [sigma_xy, sigma_yy, sigma_yz],
    [sigma_xz, sigma_yz, sigma_zz]
])

print("Tensor de Tensões (simétrico):")
sym.pprint(sigma_sym)
```
